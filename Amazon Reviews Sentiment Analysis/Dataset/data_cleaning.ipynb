```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("amazon_reviews.csv")
df.drop(["Unnamed: 0"], axis=1, inplace=True)
df
```

```python
df.dropna(subset=['reviewText'], inplace=True)
```

```python
df['reviewText'].isna().any()
```

```python
# nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
new_df = df[['reviewText']]
new_df
```

```python
transformed_df = new_df
```

### Removing StopWords

```python
transformed_df['reviewText'] = new_df['reviewText'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
transformed_df
```

### Removing Punctuations

```python
import string
transformed_df['reviewText'] = transformed_df['reviewText'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
transformed_df
```

### Converting the string to lower Case

```python
transformed_df['reviewText'] = transformed_df['reviewText'].apply(lambda x: x.lower())
transformed_df
```

### Removing Numeric characters

```python
transformed_df['reviewText'] = transformed_df['reviewText'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
transformed_df
```

### Removing emojis

```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F""]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
transformed_df['reviewText'] = transformed_df['reviewText'].apply(lambda x: remove_emoji(x))
transformed_df
```

### Removing extra spaces

```python
transformed_df['reviewText'] = transformed_df['reviewText'].apply(lambda x: ' '.join(x.strip().split()))
transformed_df
```

```python
final_df = df.replace(df['reviewText'].tolist(), transformed_df['reviewText'].tolist())
final_df.shape[0]
```

```python
final_df.to_csv('amazon_reviews_cleaned.csv')
```
